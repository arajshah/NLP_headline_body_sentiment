```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from pathlib import Path

sns.set_style("whitegrid")
Path("results").mkdir(exist_ok=True)

df = pd.read_csv("../data/full_articles_with_gap.csv")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["sent_gap"], bins=30, kde=True)
plt.axvline(0, color="black", lw=1)
plt.xlabel("Headline sentiment – Body sentiment")
plt.ylabel("Article count")
plt.title("Distribution of Sentiment Gaps")
plt.tight_layout()
plt.savefig("results/hist_gap.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y=df["sent_gap"])
plt.ylabel("Headline – Body sentiment")
plt.title("Median & Outliers")
plt.tight_layout()
plt.savefig("results/box_gap.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="sent_body", y="sent_head", data=df, alpha=0.4, s=20)
plt.axline((0,-1), slope=1, color="red", ls="--", lw=1)   # 45° reference
plt.xlabel("Body sentiment")
plt.ylabel("Headline sentiment")
plt.title("Headline vs. Body Sentiment")
plt.tight_layout()
plt.savefig("results/scatter_head_body.png", dpi=300)
plt.show()
```
